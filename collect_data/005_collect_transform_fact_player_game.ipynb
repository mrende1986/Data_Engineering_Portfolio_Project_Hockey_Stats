```python
import requests
import pandas as pd
pd.options.display.max_columns = 40
```

```python
result = pd.DataFrame()
for i in range (2020,2024):
    year = str(i)
    print(year)
    source = requests.get('https://www.hockey-reference.com/players/h/hugheja03/gamelog/'+year+'#gamelog').text
    
    df = pd.read_html(source,header=1)[0]
    df['year'] = year
    result = result.append(df, sort=False)

result.columns = ["Rk","Date","Game","Age","Tm","HA","Opp","WL","Goal","Assist","PTS","plus_minus","PIM","EVen","PPlay","SHand","GW","EV","PP","SH","S","S_pct","SHFT","TOI","HIT","BLK","FOW","FOL","FO_pct","year"]
result = result[result.Rk != 'Rk']

# save dataframe to local drive
local_filename = '/Users/mattrende/Documents/Coding/hockey-reference/YouTube Project/csv/in/jack_hughes_2020_2023.csv'
result.to_csv(local_filename, index=False)
print("CSV Saved")

```

```python
df = pd.read_csv('/Users/mattrende/Documents/Coding/hockey-reference/YouTube Project/csv/in/jack_hughes_2020_2023.csv')
df.head()
```

```python
team_abv = 'NJD'
season = df.year
reg_or_playoffs = 'R'
tracked_player_name = 'Hughes, Jack'
```

```python
df.HA = [0 if x == "@" else 1 for x in df.HA]
```

```python
df['Team_Abv'] = team_abv
df['Season'] = reg_or_playoffs
df['Season_Year'] = season
df['tracked_player_name'] = tracked_player_name
```

```python
df.head()
```

```python
abv_team_names = pd.read_csv("/Users/mattrende/Documents/Coding/hockey-reference/YouTube Project/csv/in/team_abv.csv")
abv_team_names.head(3)
```

```python
df = df.merge(abv_team_names, left_on=['Team_Abv'],right_on=["ABV"], how='left')
df.head(2)
```

```python
df.rename(columns={'ID':"team_id"},inplace=True)
```

```python
df.head()
```

```python
df["Key"] = df["team_id"].astype(str) + "-" + season.astype(str) + "-" + reg_or_playoffs + df["Game"].astype(str)
```

```python
df['Date'] = df.Date.astype(str)
df['Date'] = df.Date.str.replace('-','')
```
