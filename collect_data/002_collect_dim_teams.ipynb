```python
import pandas as pd
```

## Extract and view data

```python
url = 'https://en.wikipedia.org/wiki/National_Hockey_League'
df = pd.read_html(url)[2]
```

```python
df.head(10)
```

## Transform data

##### Drop second header row

```python
df.columns = df.columns.droplevel(-1)
df.head(2)
```

##### Create and populate a column for Conference

```python
df['Conference'] = "None"
df.head(2)
```

```python
df['Conference'][:16] = "Eastern"
df['Conference'][17:] = "Western"
```

```python
df.loc[16]
```

```python
df = df.drop(16).reset_index(drop=True)
df
```

```python
## Founded and Joined
df.Founded = df.Founded.str.replace("*","")
df.Joined = df.Joined.str.replace("*","")
```

```python
abv_team_names = pd.read_csv("/Users/mattrende/Documents/Coding/hockey-reference/YouTube Project/csv/in/team_abv.csv")
abv_team_names.head(3)
```

```python
df = df.merge(abv_team_names, on='Team', how='left')
df.head(2)
```

```python
df.columns
```

```python
df = df[['ID','Division', 'Team', 'City', 'Arena', 'Capacity', 'Founded', 'Joined',
       'General manager', 'Head coach', 'Captain', 'Conference', 'ABV']]
```

#### Export to CSV
