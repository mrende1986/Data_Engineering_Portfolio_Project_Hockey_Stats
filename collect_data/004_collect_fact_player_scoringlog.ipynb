```python
import pandas as pd
pd.options.display.max_columns = 40
```

```python
df = pd.read_csv('/Users/mattrende/Documents/Coding/hockey-reference/YouTube Project/csv/out/fact_jack_2023_scoringlog.csv')

player_name = 'Hughes, Jack'
```

```python
df.head()
```

```python
df.rename(columns={'Unnamed: 3':'Home','Unnamed: 7':'Goal_Strength'},inplace=True)
```

```python
df['Date'] = df.Date.astype(str)
df['Date'] = df.Date.str.replace('-','')
df['Date'] = df.Date.astype(int)
```

```python
## Lookup game number based on date
team_games = pd.read_csv("/Users/mattrende/Documents/Coding/hockey-reference/YouTube Project/csv/out/fact_gamelog_2023.csv")
team_games.head(3)
```

```python
df = df.merge(team_games, on=['Date'], how='left')
df.head(2)
```

```python
df['Description'][0]
```

```python
for desc in df['Description']:
    temp = desc.split(' ')[2:4]
    print(', '.join(reversed(temp)))
```

```python
goal_scorer = []
primary_assist = []
secondary_assist = []

for desc in df['Description']:

    ## Populate Goal Scorer
    temp_gs = desc.split(' ')[2:4]
    temp_gs = ', '.join(reversed(temp_gs))
    goal_scorer.append(temp_gs)

    ## Populate Primary Assist
    try:
        temp_pa = desc.split(' ')[6:8]
        temp_pa = ', '.join(reversed(temp_pa))
        primary_assist.append(temp_pa)
    except:
        primary_assist.append(None)

    ## Populate Secondary Assist
    try:
        temp_sa = desc.split(' ')[9:11]
        temp_sa = ', '.join(reversed(temp_sa))
        secondary_assist.append(temp_sa)
    except:
        secondary_assist.append(None)

df['goal_scorer'] = goal_scorer
df['primary_assist'] = primary_assist
df['secondary_assist'] = secondary_assist
df['tracked_player_name'] = player_name
```

```python
df.head(3)
```

```python
df.rename(columns={'Key':'Team_Game_Key', 'Rk':'Point_Number'},inplace=True)
df.head(1)
```

#### Reorder columns

```python
df = df[['tracked_player_name','Point_Number', 'Team_Game_Key', 'P', 'Time', 'Goal_Strength',
       'Description', 'goal_scorer', 'primary_assist', 'secondary_assist']]
```
