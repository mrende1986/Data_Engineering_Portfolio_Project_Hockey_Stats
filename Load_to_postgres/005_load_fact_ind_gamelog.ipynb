```python
import psycopg2
import pandas as pd
pd.options.display.max_columns = 40
```

### Connect to the database

```python
try:
    conn = psycopg2.connect("host=127.0.0.1 dbname=hockey_yt_project user=postgres password=password")
except psycopg2.Error as e:
    print("Error: Could not connect to database")
    print(e)

try:
    cur = conn.cursor()
except psycopg2.Error as e:
    print("Error: Could not get cursor to the database")
    print(e)

conn.set_session(autocommit=True)
```

#### create table for fact_scoring_log

```python
try:
    cur.execute("""CREATE TABLE IF NOT EXISTS fact_player_gamelog (\
                rk serial primary key, \
                key varchar,\
                Game int,\
                Age varchar,\
                Tm varchar,\
                HA varchar,\
                Opp varchar,\
                WL varchar,\
                Goal int,\
                Assist int,\
                PTS int,\
                plus_minus varchar,\
                PIM varchar,\
                EVen varchar,\
                PPlay varchar,\
                SHand varchar,\
                GW varchar,\
                EV varchar,\
                PP varchar,\
                SH varchar,\
                S varchar,\
                S_pct varchar,\
                SHFT varchar,\
                TOI varchar,\
                HIT varchar,\
                BLK varchar,\
                FOW varchar,\
                FOL varchar,\
                FO_pct varchar,\
                season_year varchar,\
                team_abv varchar,\
                Season varchar,\
                tracked_player_name varchar);""")
    
except psycopg2.Error as e:
    print("Error: Could not create table")
    print(e)
```

```python
gamelog_insert = ("""INSERT INTO fact_player_gamelog(
                key,
                Game,
                Age,
                Tm,
                HA,
                Opp,
                WL,
                Goal,
                Assist,
                PTS,
                plus_minus,
                PIM,
                EVen,
                PPlay,
                SHand,
                GW,
                EV,
                PP,
                SH,
                S,
                S_pct,
                SHFT,
                TOI,
                HIT,
                BLK,
                FOW,
                FOL,
                FO_pct,
                season_year,
                team_abv,
                Season,
                tracked_player_name)
               VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, \
                        %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, \
                        %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, \
                        %s, %s)
               """)
```

```python
## load DF for dates
df_gamelog = pd.read_csv("/Users/mattrende/Documents/Coding/hockey-reference/YouTube Project/csv/out/jack_hughes_2020_2023.csv")
df_gamelog.head(2)
```

```python
try:
    for i, row in df_gamelog.iterrows():
        #print(list(row))
        cur.execute(gamelog_insert, list(row))

except psycopg2.Error as e:
    print("Error: could not insert rows")
    print(e)
```
